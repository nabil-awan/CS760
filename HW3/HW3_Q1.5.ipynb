```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
```

```python
TPR = [0, 2/6, 4/6, 6/6, 6/6]
FPR = [0, 0/4, 1/4, 2/4, 4/4]
```

```python
TPR = np.array(TPR)
FPR = np.array(FPR)
```

```python
plt.plot(FPR, TPR, marker='o')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC plot')
plt.show()
```
